### Access sample data (phenotypic data)

```python
metadata = gse.gsms
```

### Convert metadata into a dataframe

```python
metadata_df = pd.DataFrame({gsm: metadata[gsm].metadata for gsm in metadata}).T
metadata_df
```

```python
# Reset index and rename it as 'samples' to match with expression data later
metadata_df.reset_index(inplace=True)
metadata_df.rename(columns={"index": "samples"}, inplace=True)
metadata_df.head()
```

```python
metadata_df.columns
```

### Seperate tissue and metastasis into different columns

```python
def extract_metadata(info):
    if isinstance(info, list):  # convert list to string
        info = "; ".join(info)
    parts = dict(item.strip().split(": ") for item in info.split(";") if ": " in item)
    return pd.Series({
        "tissue": parts.get("tissue", ""),
        "metastasis": parts.get("metastasis", "")
    })

# Apply the function to the 'tissue metastasis' column
metadata_split = metadata_df["characteristics_ch1"].apply(extract_metadata)

# Combine new columns with the original dataframe
#data_long = pd.concat([data_long, metadata_split], axis=1)
metadata_split
```

```python
metadata = pd.concat([metadata_df, metadata_split], axis=1)
metadata.head()
```

### Final metadata

```python
# Select and clean relevant columns
metadata = metadata[["samples", "tissue", "metastasis", "description"]].copy()
metadata
```

```python
# Remove brackets and quotes from string like "['GSM123456']"
#metadata["description"] = metadata["description"].apply(
 #   lambda x: x.strip("[]'\" ") if isinstance(x, str) else x
#)
```

```python
def extract_first(val):
    if isinstance(val, list):
        return val[0] if val else None
    return val

metadata["description"] = metadata["description"].apply(extract_first)
```

```python
metadata
```

## Load gene expression data

```python
data
```

```python
# Reshape data: wide to long format
data_long = data.rename(columns={data.columns[0]: "gene"})  # rename first column to 'gene'
data_long = pd.melt(data_long, id_vars="gene", var_name="samples", value_name="FPKM")
data_long
```

## Final dataset

```python
final_dataset = data_long.merge(metadata, left_on="samples", right_on="description", how="left")
final_dataset
```

```python
# Drop the 'description' column
final_dataset = final_dataset.drop(columns=["samples_y", "description"])
final_dataset
```

## Filter for genes of interest (BRCA1, BRCA2) and calculate basic statistics

```python
# Filter for genes of interest
filtered_df = final_dataset[final_dataset["gene"].isin(["BRCA1"])]

# Group and calculate summary statistics
summary = filtered_df.groupby(["gene", "tissue", "metastasis"]).agg(
    mean_FPKM=("FPKM", "mean"),
    median_FPKM=("FPKM", "median"),
    std_FPKM=("FPKM", "std")
).reset_index()

print(summary)
```

## Visualization

### BRCA1 expression across samples

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for BRCA1
brca1_data = final_dataset[final_dataset["gene"] == "BRCA1"]

# Set the plot size
plt.figure(figsize=(12, 6))

# Create the barplot
sns.barplot(data=brca1_data, x="samples_x", y="FPKM", hue="tissue")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# Add labels and title
plt.title("BRCA1 Expression (FPKM) Across Samples")
plt.xlabel("Sample")
plt.ylabel("FPKM")

# Show legend and plot
plt.legend(title="Tissue")
plt.tight_layout()
plt.show()
```

#### t-test

```python
from scipy.stats import ttest_ind

# Filter BRCA1 data (you already did this)
brca1_data = final_dataset[final_dataset["gene"] == "BRCA1"]

# Select two tissues to compare, for example: "Tumor" vs "Normal"
tumor_expr = brca1_data[brca1_data["tissue"] == "breast tumor"]["FPKM"]
normal_expr = brca1_data[brca1_data["tissue"] == "normal breast tissue"]["FPKM"]

# Perform independent t-test (Welch’s t-test recommended if variances are unequal)
t_stat, p_value = ttest_ind(tumor_expr, normal_expr, equal_var=False)

# Print the results
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Optional: interpret result
if p_value < 0.05:
    print("Significant difference in BRCA1 expression between Tumor and Normal tissues.")
else:
    print("No significant difference in BRCA1 expression between Tumor and Normal tissues.")
```

### Density plot of BRCA1 expression by tissue

```python
from scipy.stats import mannwhitneyu

# Filter BRCA1 data
brca1_data = final_dataset[final_dataset["gene"] == "BRCA1"]

# Select two tissues to compare, e.g., "Tumor" vs "Normal"
tumor_expr = brca1_data[brca1_data["tissue"] == "breast tumor"]["FPKM"]
normal_expr = brca1_data[brca1_data["tissue"] == "normal breast tissue"]["FPKM"]

# Perform Mann–Whitney U test
u_stat, p_value = mannwhitneyu(tumor_expr, normal_expr, alternative='two-sided')

# Print the result
print(f"U-statistic: {u_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Optional interpretation
if p_value < 0.05:
    print("Significant difference in BRCA1 expression distribution between Tumor and Normal tissues.")
else:
    print("No significant difference in BRCA1 expression distribution between Tumor and Normal tissues.")
```

### 3. Violin plot of BRCA1 expression by metastasis status

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=brca1_data, x='metastasis', y='FPKM')
plt.title('BRCA1 Expression by Metastasis Status')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Mann–Whitney U Test (non-parametric)

```python
from scipy.stats import mannwhitneyu

# Grouped data
group1 = brca1_data[brca1_data["metastasis"] == "yes"]["FPKM"]
group2 = brca1_data[brca1_data["metastasis"] == "no"]["FPKM"]

# Mann–Whitney U test
u_stat, p_value = mannwhitneyu(group1, group2, alternative='two-sided')

print(f"Mann–Whitney U-statistic: {u_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant difference in BRCA1 expression between Metastatic and Non-Metastatic groups.")
else:
    print("No significant difference in BRCA1 expression between Metastatic and Non-Metastatic groups.")
```

#### t-test (parametric)

```python
from scipy.stats import ttest_ind

# Grouped data
group1 = brca1_data[brca1_data["metastasis"] == "yes"]["FPKM"]
group2 = brca1_data[brca1_data["metastasis"] == "no"]["FPKM"]

# T-test (Welch’s)
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant difference in BRCA1 expression between Metastatic and Non-Metastatic groups.")
else:
    print("No significant difference in BRCA1 expression between Metastatic and Non-Metastatic groups.")
```

### Scatter plot of BRCA1 vs BRCA2 expression

```python
brca_data = final_dataset[final_dataset['gene'].isin(['BRCA1', 'BRCA2'])]
brca_pivot = brca_data.pivot_table(index=['samples_x', 'tissue'],
                                    columns='gene',
                                    values='FPKM').reset_index()
plt.figure(figsize=(8, 6))
sns.scatterplot(data=brca_pivot, x='BRCA1', y='BRCA2', hue='tissue')
sns.regplot(data=brca_pivot, x='BRCA1', y='BRCA2', scatter=False, color='black')
plt.title('BRCA1 vs BRCA2 Expression')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Spearman Correlation for BRCA1 vs BRCA2 Expression

```python
from scipy.stats import spearmanr

# Filter data for BRCA1 and BRCA2
brca1_expr = final_dataset[final_dataset["gene"] == "BRCA1"][["samples_x", "FPKM"]].rename(columns={"FPKM": "BRCA1_FPKM"})
brca2_expr = final_dataset[final_dataset["gene"] == "BRCA2"][["samples_x", "FPKM"]].rename(columns={"FPKM": "BRCA2_FPKM"})

# Merge on sample ID
merged = pd.merge(brca1_expr, brca2_expr, on="samples_x")

# Calculate Spearman correlation
corr, p_value = spearmanr(merged["BRCA1_FPKM"], merged["BRCA2_FPKM"])

# Print results
print(f"Spearman correlation coefficient: {corr:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Significant monotonic relationship between BRCA1 and BRCA2 expression.")
else:
    print("No significant monotonic relationship between BRCA1 and BRCA2 expression.")
```
